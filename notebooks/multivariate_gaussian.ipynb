```python
des = train_df.describe().T
gaussian_model = []
for i in range(1, 31) :
    index = 'V' + str(i)
    gaussian_model.append([des.loc[index, 'mean'], des.loc[index, 'std']])
```

```python
gaussian_model[0][0]
```

```python
val_df.head()
```

```python

col=[col for col in val_df.columns if col not in ['ID','Class']]
val_prob = pd.DataFrame(columns=val_df.columns)

for i, v in enumerate(col):
    val_prob[v] = val_df[v].apply(lambda x : scipy.stats.norm(gaussian_model[i][0], gaussian_model[i][1]).pdf(x))
```

```python
val_prob.head()
```

```python
val_prob['ID'] = val_df['ID'].values
val_prob['Class'] = val_df['Class'].values
val_prob.head()
```

```python
import numpy as np

def geo_mean_overflow(iterable):
    # print(iterable)
    return np.log(np.exp(iterable).mean())
```

```python
val_prob['prob'] = val_prob.apply(lambda x : geo_mean_overflow(x[1:-1]), axis=1)
```

```python
val_prob.head()
```

```python
# np.log(np.exp(val_prob.iloc[0, :][1:-1].values).mean())
```

```python
import torch
topk_indices = torch.topk(torch.tensor(val_prob['prob'].values), k=100, largest=False).indices
```

```python
pred_y = np.zeros(len(val_prob))
pred_y[topk_indices] = 1
```

```python
np.where(pred_y != val_prob['Class'].values)
```

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
```

```python
val_score = f1_score(val_prob['Class'].values, pred_y, average='macro')
val_score
```
